\begin{align*}

\begin{bmatrix}
2&3&1&1&4\\
1&2&2&3&1\\

1&1&1&2&2\\
1&3&2&1&4
\end{bmatrix}
\end{align*}

untuk menghitung invers kita mengunakan rumus:

$$ A^{-1} = \frac{1}{\text{det}(A)} \times \text{adj}(A) $$

sedangkan

$$ \text{adj}(A)={\text({conf}(A))^T}$$

untuk

$${\text{det}(A)}=a_{i,j}*c_{i,j}+a_{i,j}*c_{i,j}+...+a_{i,j}*c_{i,j}$$

dengan

$$c_{i,j}=(-1)^{i+j}*det(a_{i,j})$$
