# MatrikTanformasi

Transformasi pada matriks adalah proses di mana kita menerapkan operasi tertentu pada matriks untuk mengubahnya menjadi bentuk lain. Ada beberapa jenis transformasi pada matriks yang sering digunakan dalam berbagai aplikasi matematika, seperti aljabar linier, grafik komputer, dan analisis data. Berikut adalah beberapa penjelasan tentang transformasi umum pada matriks:






##1. **Refleksi (Reflection)**:
   - Refleksi adalah transformasi yang mencerminkan objek terhadap sumbu atau garis tertentu.
   - Misalnya, refleksi terhadap sumbu y di ruang 2D diberikan oleh matriks:
     \[
     R = \begin{bmatrix}
     -1 & 0 \\
     0 & 1
     \end{bmatrix}
     \]

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan direfleksi)
a = np.array([[2, 6]])


# Definisikan matriks refleksi
R = np.array([[-1, 0],
              [0, 1]])

# Hitung hasil refleksi
c = a @ R
print("Hasil refleksi matrik a pada sumbu y", c)

# Visualisasi transformasi vektor sebelum dan sesudah refleksi
plt.figure(figsize=(12, 6))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor asli (np1)')

# Vektor setelah refleksi
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Vektor rotasi (c)')

# Pengaturan plot
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Refleksi Vektor oleh Matriks')
# Refleksi visualisasi:
# Vektor asli (a) = [4, 4] ditampilkan dalam warna biru.
# Vektor setelah refleksi (c) = hasil rotasi a sebesar 45 derajat ditampilkan dalam warna hijau.
# Refleksi ini mengubah arah vektor a terhadap sumbu y.

# Menampilkan grafik
plt.show()
```

##2. **Rotasi (Rotation)**:
   - Rotasi adalah jenis transformasi yang memutar vektor atau titik di sekitar asal koordinat (0,0) di ruang 2D atau di sekitar sumbu di ruang 3D.
   - Untuk rotasi di ruang 2D dengan sudut \(theta \), matriks rotasinya adalah:
     \[
     R(theta) = \begin{bmatrix}
     \cos\theta & -\sin\theta \\
     \sin\theta & \cos\theta
     \end{bmatrix}
     \]
   - Di ruang 3D, rotasi bisa dilakukan sekitar sumbu x, y, atau z dengan matriks rotasi yang sesuai.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan dirotasi)
a = np.array([[4, 4]])

# Definisikan sudut rotasi (dalam derajat) dan konversi ke radian
theta = 90  # Sudut rotasi dalam derajat
theta_rad = np.radians(theta)  # Konversi ke radian

# Definisikan matriks rotasi
R = np.array([[np.cos(theta_rad), -np.sin(theta_rad)],
              [np.sin(theta_rad), np.cos(theta_rad)]])

# Hitung hasil rotasi
c = a @ R
print("Hasil rotasi matriks a dengan sudut", theta, "derajat adalah:", c)
# Output: Hasil rotasi matriks a dengan sudut 90 derajat adalah: [[0. 5.65685425]]

# Visualisasi transformasi vektor sebelum dan sesudah rotasi
plt.figure(figsize=(12, 6))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor asli (a)')

# Vektor setelah rotasi
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Vektor rotasi (c)')

# Pengaturan plot
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Rotasi Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [4, 4] ditampilkan dalam warna biru.
# Vektor setelah rotasi (c) = hasil rotasi a sebesar 45 derajat ditampilkan dalam warna hijau.
# Rotasi ini mengubah arah vektor a sesuai dengan sudut yang diberikan.

# Menampilkan grafik
plt.show()
```

##3. **Shear (Penggeseran)**:
   - Shear adalah transformasi yang menggeser titik-titik dalam satu arah (x atau y) tergantung pada posisi mereka dalam arah yang lain.
   - Contoh matriks shear dalam ruang 2D adalah:
     \[
     \text{Shear} = \begin{bmatrix}
     1 & k \\
     0 & 1
     \end{bmatrix}
     \]
     di mana \( k \) adalah faktor shear.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan shear)
a = np.array([[2, 3]])

# Definisikan faktor shear untuk melakukan pergeseran matrik
k = 2 #faktor shear

# Definisikan matriks shear horizontal pada sumbu y
R = np.array([[1,k],
              [0,1]])

# Hitung hasil pergeseran
c = a @ R
print("Hasil sheaar matriks a dengan faktor shear", k, "horizontal pada sumbu y", c)


# Visualisasi transformasi vektor sebelum dan sesudah pergeseran
plt.figure(figsize=(12, 6))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor asli (a)')

# Vektor setelah pergeseran
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Vektor rotasi (c)')

# Pengaturan plot
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi shear Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [2, 3] ditampilkan dalam warna biru.
# Vektor setelah pergeseran (c) = hasil rotasi a dalam warna hijau.
# pergeseran ini mengubah arah vektor a sesuai dengan faktor shear 2.

# Menampilkan grafik
plt.show()
```

##2. **Skalasi (Scaling)**:
   - Skalasi mengubah ukuran objek dengan memperbesar atau memperkecilnya.
   - Matriks skalasi untuk ruang 2D dengan faktor skalasi \( s_x \) dan \( s_y \) adalah:
     \[
     S = \begin{bmatrix}
     s_x & 0 \\
     0 & s_y
     \end{bmatrix}
     \]
   - Untuk ruang 3D, faktor skalasi juga dapat diterapkan pada sumbu z.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan diskalasi)
a = np.array([[2, 1]])

# Definisikan faktor skalasi x dan faktor skalasi y  untuk melakukan scaling matrik
fx = 2 #faktor skalasi x
fy = 3 #faktor skalasi y

# Definisikan matriks shear horizontal pada sumbu y
R = np.array([[fx,0],
              [0,fy]])

# Hitung hasil skalasi
c = a @ R
print("Hasil scaling matriks a dengan faktor x", fx, "dan fakto y", fy, "adalah", c)


# Visualisasi transformasi vektor sebelum dan sesudah skalasi
plt.figure(figsize=(12, 6))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor asli (a)')

# Vektor setelah scaling
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Vektor rotasi (c)')

# Pengaturan plot
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi scaling Vektor oleh Matriks')

# scaling visualisasi:
# Vektor asli (a) = [2, 1] ditampilkan dalam warna biru.
# Vektor setelah scaling (c) = hasil rotasi a dalam warna hijau.
# pergeseran ini mengubah arah vektor a sesuai dengan faktor x 2 dan faktor y 3.

# Menampilkan grafik
plt.show()
```

Setiap transformasi ini dapat direpresentasikan dengan matriks, dan menerapkan transformasi pada vektor berarti mengalikan vektor tersebut dengan matriks transformasi yang sesuai. Dalam aplikasi praktis, seringkali beberapa transformasi digabungkan menjadi satu dengan mengalikan matriks-matriks transformasi yang berbeda.
